### importing libraries

```python
import pandas as pd
import numpy as np
```

### Data ingestion

```python
#insert data in spark
data_path= spark.read.table('workspace.default.retail_sales_dataset') 

#convert .toPandas
df=data_path.toPandas()

# show/display table 
display(df)
```

### questions

```python
# 1. Display all columns for all transactions. 

df.columns
```

```python
# 2. Display only the Transaction ID, Date, and Customer ID for all records. 

df[['Transaction ID', 'Date', 'Customer ID']]
```

```python
# Q3. Display all the distinct product categories in the dataset.

df['Product Category'].unique() 
```

```python
# Q4. Display all the distinct gender values in the dataset. 

df['Gender'].unique()
```

```python
# Q5. Display all transactions where the Age is greater than 40. 

over_40_yrs_age=df[(df['Age'] > 40)].sort_values(by ='Age',ascending=True)



display(over_40_yrs_age)
```

```python
# Q6. Display all transactions where the Price per Unit is between 100 and 500. 


price_range=df[df['Price per Unit'].between(100,500)].sort_values('Price per Unit')

display(price_range)
```

```python
# Q7. Display all transactions where the Product Category is either 'Beauty' or 'Electronics'. 

df[(df['Product Category'] == 'Beauty') | (df['Product Category'] == 'Electronics')]
```

```python
# Q8. Display all transactions where the Product Category is not 'Clothing'. 

not_Clothing_product=df[(df['Product Category'] != 'Clothing')]

display(not_Clothing_product)
```

```python
# Q9. Display all transactions where the Quantity is greater than or equal to 3. 

more_than_2_quantity= df[(df['Quantity'] >= 3)]

display(more_than_2_quantity)
```

```python
# Q10. Count the total number of transactions. 

len(df)

df.shape
```

```python
# Q11. Find the average Age of customers. 

avg_age = df.agg(average_age=('Age', 'mean')).round(0).astype(int)

display(avg_age)
```

```python
# Q12. Find the total quantity of products sold. 

product_quantity_sold=df.groupby('Product Category',as_index=False).agg(total_quantity =('Quantity','sum'))

display(product_quantity_sold)
```

```python
# Q13. Find the maximum Total Amount spent in a single transaction. 

max_total_amount= df.agg(max_amount=('Total Amount','max'))

display(max_total_amount)
```

```python
# Q14. Find the minimum Price per Unit in the dataset. 

min_price_unit= df.agg(min_price=('Price per Unit','min'))

display(min_price_unit)
```

```python
# Q15. Find the number of transactions per Product Category

transactions per Product Category=df.groupby('Product Category',as_index=False).agg(transaction_count=('Transaction ID','count'))

display(transactions per Product Category)
```

```python
y# Q16. Find the total revenue (Total Amount) per gender. 

Total_Amount_per_gender=df.groupby('Gender', as_index=False).agg(total_revenue=('Total Amount','sum'))

display( Total_Amount_per_gender)

```

```python
# 17. Find the average Price per Unit per product category. 

average_price_per_product=df.groupby('Product Category',as_index=False).agg(avg_unit_price=('Price per Unit','mean')).round(0)

display(average_price_per_product)
```

```python
# Q18. Find the total revenue per product category where total revenue is greater than 10,000.

greater_than_10k_product=(df.groupby('Product Category',as_index=False).agg(total_revenue=('Total Amount','sum')).query('total_revenue > 10000'))

display(greater_than_10k_product)
```

```python
# Q19. Find the average quantity per product category where the average is more than 2. 

avg_quantity_product=(df.groupby('Product Category',as_index=False).agg(avg_quantity=('Quantity','mean')).round(0).query('avg_quantity > 2'))


display(avg_quantity_product)
```

```python
# Q20. Display a column called Spending_Level that shows 'High' if Total Amount > 1000, otherwise 'Low'.


condition = [df['Total Amount'] > 1000]

choice = ['High']

df['Spending_Level'] = np.select(condition, choice, default='Low')

display(df[['Total Amount', 'Spending_Level']].head())
```

```python
# Q21. Display a new column called Age_Group that labels customers as:

basket=[(df['Age']< 30),(df['Age'].between(30,59))]

group_12 =['youth', 'Adults']

df['basket_group'] = np.select(basket, group_12, default='senior')

display(df[['Age', 'basket_group']].head(5))
```
